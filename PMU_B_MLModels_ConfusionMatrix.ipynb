```python
# @title Confusion Matrix from train Logistic regression with 10 iteration

from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_iris  # Import load_iris
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Iris dataset
iris = load_iris()
X, y = iris.data, iris.target # Use the data and target from the dataset

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train a simple model (replace with your trained model)
model_10 = LogisticRegression(max_iter=10) # Increased max_iter for convergence
model_10.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model_10.predict(X_test)

# Generate the confusion matrix
cm_10 = confusion_matrix(y_test, y_pred)

# Display the confusion matrix using seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(cm_10, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=iris.target_names, yticklabels=iris.target_names) # Add labels
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix (Iris Dataset)')
plt.show()
```

```python
# @title Calculate evaluation matrices (without using library)

def calculate_metrics(cm):
    """Calculates precision, sensitivity, specificity, F1-score, and accuracy for a multi-class confusion matrix."""
    n_classes = cm.shape[0]
    accuracy = np.trace(cm) / np.sum(cm)
    precision = np.zeros(n_classes)
    sensitivity = np.zeros(n_classes) # Also known as recall
    specificity = np.zeros(n_classes)
    f1_score = np.zeros(n_classes)

    for i in range(n_classes):
        tp = cm[i, i]
        fp = np.sum(cm[:, i]) - tp
        fn = np.sum(cm[i, :]) - tp
        tn = np.sum(cm) - tp - fp - fn

        precision[i] = tp / (tp + fp) if (tp + fp) > 0 else 0
        sensitivity[i] = tp / (tp + fn) if (tp + fn) > 0 else 0
        specificity[i] = tn / (tn + fp) if (tn + fp) > 0 else 0
        f1_score[i] = 2 * (precision[i] * sensitivity[i]) / (precision[i] + sensitivity[i]) if (precision[i] + sensitivity[i]) > 0 else 0

    # Calculate macro average
    macro_precision = np.mean(precision)
    macro_sensitivity = np.mean(sensitivity)
    macro_specificity = np.mean(specificity)
    macro_f1_score = np.mean(f1_score)

    return accuracy, macro_precision, macro_sensitivity, macro_specificity, macro_f1_score

# Assuming cm_5 and cm_10 are already generated from previous cells
# (Confusion matrices for max_iter=5 and max_iter=10)

# Calculate metrics for max_iter=5
accuracy_5, precision_5, sensitivity_5, specificity_5, f1_score_5 = calculate_metrics(cm_5)

print("Metrics for max_iter=5:")
print(f"Accuracy: {accuracy_5:.4f}")
print(f"Macro Precision: {precision_5:.4f}")
print(f"Macro Sensitivity (Recall): {sensitivity_5:.4f}")
print(f"Macro Specificity: {specificity_5:.4f}")
print(f"Macro F1-score: {f1_score_5:.4f}")

print("-" * 20)

# Calculate metrics for max_iter=10
accuracy_10, precision_10, sensitivity_10, specificity_10, f1_score_10 = calculate_metrics(cm_10)

print("Metrics for max_iter=10:")
print(f"Accuracy: {accuracy_10:.4f}")
print(f"Macro Precision: {precision_10:.4f}")
print(f"Macro Sensitivity (Recall): {sensitivity_10:.4f}")
print(f"Macro Specificity: {specificity_10:.4f}")
print(f"Macro F1-score: {f1_score_10:.4f}")
```

```python
# @title Calculate evaluation matrices through sklearn library (classification_report)
from sklearn.metrics import classification_report

# Assuming model_5 and model_10 are the trained models from previous cells

# Classification report for max_iter=5
y_pred_5 = model_5.predict(X_test)
report_5 = classification_report(y_test, y_pred_5, target_names=iris.target_names)
print("Classification Report for max_iter=5:")
print(report_5)

print("-" * 30)

# Classification report for max_iter=10
y_pred_10 = model_10.predict(X_test)
report_10 = classification_report(y_test, y_pred_10, target_names=iris.target_names)
print("Classification Report for max_iter=10:")
print(report_10)
```

```python
# @title Calculate ROC-AUC with using sklearn library
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize

# Assuming you have the trained models (model_5 and model_10) from previous cells
# and the test data (X_test, y_test)

# Binarize the output for one-vs-rest ROC
y_test_bin = label_binarize(y_test, classes=[0, 1, 2])  # class: ['setosa', 'versicolor', 'verginica']
n_classes = y_test_bin.shape[1]

# Calculate ROC and AUC for max_iter=5
y_score_5 = model_5.predict_proba(X_test)
fpr_5 = dict()
tpr_5 = dict()
roc_auc_5 = dict()
for i in range(n_classes):
    fpr_5[i], tpr_5[i], _ = roc_curve(y_test_bin[:, i], y_score_5[:, i])
    roc_auc_5[i] = auc(fpr_5[i], tpr_5[i])

# Calculate micro-average ROC curve and AUC
fpr_5["micro"], tpr_5["micro"], _ = roc_curve(y_test_bin.ravel(), y_score_5.ravel())
roc_auc_5["micro"] = auc(fpr_5["micro"], tpr_5["micro"])

# Calculate ROC and AUC for max_iter=10
y_score_10 = model_10.predict_proba(X_test) # Assuming model with max_iter=10 is still assigned to 'model'
fpr_10 = dict()
tpr_10 = dict()
roc_auc_10 = dict()
for i in range(n_classes):
    fpr_10[i], tpr_10[i], _ = roc_curve(y_test_bin[:, i], y_score_10[:, i])
    roc_auc_10[i] = auc(fpr_10[i], tpr_10[i])

# Calculate micro-average ROC curve and AUC
fpr_10["micro"], tpr_10["micro"], _ = roc_curve(y_test_bin.ravel(), y_score_10.ravel())
roc_auc_10["micro"] = auc(fpr_10["micro"], tpr_10["micro"])


# Plot ROC curves
plt.figure(figsize=(10, 8))

# Plot for max_iter=5
plt.plot(fpr_5["micro"], tpr_5["micro"],
         label=f'max_iter=5 (Micro-average AUC = {roc_auc_5["micro"]:.2f})',
         color='deeppink', linestyle=':', linewidth=4)

# Plot for max_iter=10
plt.plot(fpr_10["micro"], tpr_10["micro"],
         label=f'max_iter=10 (Micro-average AUC = {roc_auc_10["micro"]:.2f})',
         color='darkorange', linestyle=':', linewidth=4)


plt.plot([0, 1], [0, 1], 'k--', lw=2)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve (One-vs-Rest)')
plt.legend(loc="lower right")
plt.show()
```
